**Train Data**

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
encoder = LabelEncoder()
df['Target'] = encoder.fit_transform(df['Target'])

from sklearn.model_selection import train_test_split
X= df.drop(['Target','Target_Names'], axis="columns")
y= df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

**Min Max Standardization**

```python
from sklearn.preprocessing import MinMaxScaler

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the dataset
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn import tree

model1= tree.DecisionTreeClassifier()
model1.fit(X_train_scaled, y_train)
model1
```

```python
model1.score(X_test_scaled, y_test)
```

```python
model1.predict([[1,1,5,5,1,1,1,22,27,10,10,1,0,0,0,1,0,19,0,0,6,0,0,0,0,0,6,0,0,0,0,10.8,1.4,1.74]])
```

```python
y_predicted1= model1.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
cm= confusion_matrix(y_test, y_predicted1)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot()
plt.show()
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

classifier1= RandomForestClassifier(n_estimators= 50)
classifier1.fit(X_train_scaled, y_train)
```

```python
classifier1.score(X_test_scaled, y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics

y_predicted2= classifier1.predict(X_test_scaled)

cm= confusion_matrix(y_test, y_predicted2)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot()
plt.show()
```

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
classifier2= GaussianNB()
classifier2.fit(X_train_scaled, y_train)
```

```python
classifier2.score(X_test_scaled, y_test)
```

```python
classifier2.predict(X_test_scaled[:10])
```

```python
y_test[:10]
```

```python
y_predicted3= classifier2.predict(X_test_scaled)

cm= confusion_matrix(y_test, y_predicted3)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot()
plt.show()
```

**SVM**

```python
from sklearn.svm import SVC
model2= OneVsRestClassifier(SVC(probability=True))
model2.fit(X_train_scaled, y_train)
```

```python
model2.score(X_test_scaled, y_test)
```

```python
model2.predict(X_test_scaled[:10])
```

```python
y_test[:10]
```

```python
y_predicted4= model2.predict(X_test_scaled)

cm= confusion_matrix(y_test, y_predicted4)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot()
plt.show()
```

**Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingClassifier
gbr= GradientBoostingClassifier(n_estimators=100)
gbr.fit(X_train_scaled, y_train)
```

```python
gbr.score(X_test_scaled, y_test)
```

```python
y_predicted5= gbr.predict(X_test_scaled)

cm= confusion_matrix(y_test, y_predicted5)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot()
plt.show()
```

**AUC Score**

```python
from sklearn.metrics import auc, roc_curve

r_probs=[0 for i in range(len(y_test))]
y1_pred_prob=model1.predict_proba(X_test_scaled)
y2_pred_prob=model2.predict_proba(X_test_scaled)
y3_pred_prob=classifier1.predict_proba(X_test_scaled)
y4_pred_prob=classifier2.predict_proba(X_test_scaled)
y5_pred_prob=gbr.predict_proba(X_test_scaled)
```

```python
auc_score=[metrics.roc_auc_score(y_test,i, multi_class='ovr') for i in [y1_pred_prob, y2_pred_prob,y3_pred_prob,y4_pred_prob,y5_pred_prob] ]
```

```python
plt.figure(figsize=(14,8))
models=["Decision Tree", "SVM", "Random Forest", "Naive Bayes", "Gradient Boosting"]
plt.bar(models,auc_score,color='midnightblue',width=0.8)
plt.xlabel('Models',fontsize=20)
plt.ylabel('AUC SCORE',fontsize=20)
plt.title('COMPARISION OF AUC SCORE',fontsize=26,weight='bold')
#plt.ylim(top=1)
#plt.ylim(bottom=0.95)
plt.grid(color='b',ls='-.',lw=0.25)

#for displaying the value above plot
#plt.text(-.35,0.9982,str(auc_score[0]),fontsize=16)
#plt.text(0.65,0.9942,str(auc_score[1]),fontsize=16)
#plt.text(1.65,0.9919,str(auc_score[2]),fontsize=16)
plt.show()
```
